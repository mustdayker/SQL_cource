## 1.1 Типы операций над таблицами [▴](#head)

Посмотреть версию SQL сервера

```sql
SELECT @@Version
```

Использовать в запросе определенную БД

```sql
USE DB_name
```

В таблице нет нумерации строк. Обращаться по номерам нельзя. Таблица это не упорядоченный массиыы.

Есть два типа таблиц
- содержат списки объектов (записей)
- содержат списки связей между объектами (создаются для связей многие ко мнонгим)


Операции над таблицами можно разделить на две группы:
- простые операции
- трасформирующие

### Простые (не трансформирующие таблицу) - не меняют тип списка (таблицы)
- `WHERE`
- `SELECT`
- `ORDER BY`
- `TOP` (`LIMIT`) - первые N строк
- `DISTINCT` - устранение дубликатов строк

Порядок операций в запросе
1. `FROM`
2. `WHERE`
3. `ORDER BY` → `SELECT`, `TOP`, `DISTINCT`

### Трансформирующие операции
- Агрегация (грегатные функции)
- `GROUP BY` - переход от одного списка объектов к списку групп объектов
- `HAVING` - фильтрация групп (применяется после группировки)

## 1.2 Сведение вместе нескольких таблиц (Подзапросы) [▴](#head)

### Принципиальный выбор стратегии: подзапросы или джойны

![](data/scr_0006.jpg)

### Подзапросы
1. Список чего я хочу получить в результате
2. Написать `SELECT` из той таблицы где лежит этот список
3. В какой таблице лежит то, что мне сейчас нехватает?
4. Пишем подзапрос `SELECT` к той таблице где это есть

### Задача

![](data/scr_0007.jpg)

![](data/scr_0008.jpg)

### Плюсы подзапросов
- Простые. Работая с подзапросами не надо делать трансформации (решаем без использования `GROUP BY` и `HAVING`)
- Не надо держать в голове все таблицы (как с `JOIN`)
- Надежные (сложно ошибитсья)

### Минусы подзапросов
- Плохо поддаются оптимизации
- Быстро становятся громоздкими
- Требовательные к структуре базы

### Задача

В какие страны поставлялись приправы (`CategoryName = 'Condiments'`)

### Решение

```sql
SELECT DISTINCT ShipCountry
FROM  Orders
WHERE OrderID IN (
                  SELECT OrderID
                  FROM  [Order Details]
                  WHERE  ProductID IN (
                                       SELECT ProductID
                                       FROM   Products
                                       WHERE  CategoryID = (
                                                            SELECT CategoryID
                                                            FROM   Categories
                                                            WHERE  CategoryName = 'Condiments'
                                                           )
                                      )
                 )
```

![](data/scr_0009.jpg)

### Задача

Сколько штук (`Quantity`) мы продали по каждой категории в Германию

### Решение

```sql
SELECT CategoryName,
       (
        SELECT SUM (Quantity)
        FROM   [Order Details]
        WHERE  ProductID IN (
                             SELECT ProductID
                             FROM   Products
                             WHERE  CategoryID = Categories.CategoryID
                            )
           AND OrderID IN (
                           SELECT OrderID
                           FROM Orders
                           WHERE ShipCountry = 'Germany'
                           )
       )
FROM   Categories
```
